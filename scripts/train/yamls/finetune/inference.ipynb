```python
test_data = pd.read_csv("test.tsv", sep="\t")

raw_finetuned_host = "https://mpt-7b-8k-raw-sut-finetuned-d4adkg.inf.hosted-on.mosaicml.hosting"
test_data
```

```python
failed_indices = []
predictions_raw_finetuned = []
for i, row in test_data.iterrows():
    try:
        raw_fn_predcition = predict(raw_finetuned_host, 
                                    {
                                        "inputs": [row["input"]], 
                                        "temperature": 0.,
                                        "max_length": 8192,
                                    }, 
                                    timeout=600)
        predictions_raw_finetuned.append(raw_fn_predcition["output"][0])
        # print(raw_fn_predcition["output"][0])
        print(i)
    except:
        print(f"{i} failed")
        failed_indices.append(i)

```
