```python
test_data = pd.read_csv("test.tsv", sep="\t")

ava_finetuned_host = "https://mpt-7b-8k-ava-sut-finetuned-lbm2q0.inf.hosted-on.mosaicml.hosting"
test_data
```

```python
failed_indices = []
predictions_ava_finetuned = []
for i, row in test_data.iterrows():
    try:
        ava_fn_predcition = predict(ava_finetuned_host, 
                                    {
                                        "inputs": [row["input"]], 
                                        "temperature": 0.,
                                        "max_length": 8192,
                                    }, 
                                    timeout=600)
        predictions_ava_finetuned.append(ava_fn_predcition["output"][0])
        # print(raw_fn_predcition["output"][0])
        print(i)
    except:
        print(f"{i} failed")
        failed_indices.append(i)

```

```python
def get_labels(i, text):
    # Search for content including the [[...]]
    # match = re.search(r'(\[\[.*?\]\])', text)
    
    # if match:
    #     content_with_brackets = match.group(1)
    #     return content_with_brackets
    # else:
    #     print("No labels found!", i)
    #     return None

    res = "[[" + text.split("[[")[-1]
    if res.endswith("]]"):
        return res
    else:
        return res + "']]"

    
predictions = [get_labels(i, item) for i, item in enumerate(predictions_ava_finetuned)]
```
